```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sklearn 
from sklearn import linear_model
from sklearn.metrics import r2_score

from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
current_dir = os.getcwd()
path_file = os.path.join(current_dir, "datasets/ConsumoxCo2.csv")
data = pd.read_csv(path_file, header=0)
```

```python
#Regressão linear com mais de uma variável independente
data.describe()
```

```python
X = data[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', \
            'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG',]]
```

```python
Y = data['CO2EMISSIONS']
```

```python
selection = pd.concat([X, Y], axis=1)
selection.head()
```

```python
train = selection[:(int((len(selection)*0.8)))]
test = selection[(int((len(selection)*0.8))):]
```

```python
reg = linear_model.LinearRegression()

train_x = np.array(train[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', \
                            'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG']])
train_y = np.array(train[['CO2EMISSIONS']])
test_x = np.array(test[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', \
                            'FUELCONSUMPTION_COMB', 'FUELCONSUMPTION_COMB_MPG']])
test_y = np.array(test[['CO2EMISSIONS']])

reg.fit(train_x, train_y)
```

```python
coefs = pd.DataFrame(reg.coef_.flatten(), X.columns, columns=["Coeficientes"])
coefs
```

```python
Y_pred = reg.predict(test_x)
```

```python
R = r2_score(test_y, Y_pred)
print("R^2: ", R)
```
