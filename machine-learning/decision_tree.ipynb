```python
import pandas as pd
import os

current_dir = os.getcwd()
path_file = os.path.join(current_dir, "datasets/consumption_of_petrol.csv")
df_petr = pd.read_csv(path_file, header=0)
```

```python
df_petr.head()
```

```python
X = df_petr.drop('Petrol_Consumption', axis=1)
y = df_petr['Petrol_Consumption']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regres = DecisionTreeRegressor()
regres.fit(X_train, y_train)
```

```python
y_pred = regres.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print('Erro quadratico medio (MSE): ', mse)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(48, 32))
plot_tree(regres, filled = True)
plt.show()
```
