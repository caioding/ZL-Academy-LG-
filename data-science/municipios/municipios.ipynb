```python
# Vendo se os tipos tipos dos objetos
data = df_am_pop_
df = pd.DataFrame(data)
print(df.dtypes)
```

```python
# Limpeza da coluna ±% para remover % e tratar valores inválidos
df["±%"] = df["±%"].str.replace("%", "")  # Remove o símbolo %
df["±%"] = pd.to_numeric(df["±%"], errors="coerce")  # Converte para número, substitui inválidos por NaN

# Convertendo outras colunas
df["Year"] = df["Year"].astype(int)
df["Pop."] = df["Pop."].astype(int)
```

```python
# Tirando os NaN
df["±%"].fillna(0, inplace=True)
print(df)
```
